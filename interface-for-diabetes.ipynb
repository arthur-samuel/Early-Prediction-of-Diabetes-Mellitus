```python
import joblib
import numpy as np
```

```python
# Load the trained model
model = joblib.load('/kaggle/input/diabetes_model/keras/diabetes_model/1/Diabetes_model')
```

```python


def predict_diabetes():
    print("Welcome to the Diabetes Prediction System")
    print("Please enter the following parameters:")
    
    pregnancies = int(input("Number of Pregnancies: "))
    glucose = float(input("Glucose Level: "))
    blood_pressure = float(input("Blood Pressure (mm Hg): "))
    skin_thickness = float(input("Skin Thickness (mm): "))
    insulin = float(input("Insulin Level (mu U/ml): "))
    bmi = float(input("Body Mass Index (BMI): "))
    dpf = float(input("Diabetes Pedigree Function: "))
    age = int(input("Age (years): "))
    
    # Make prediction
    input_data = np.array([[glucose, blood_pressure, skin_thickness, insulin, bmi, dpf, age]])
    
    prediction = model.predict(input_data)
    
    
    # Convert prediction to human-readable format
    if prediction[0] == 1:
        result = 'Positive for diabetes'
    else:
        result = 'Negative for diabetes'
    
    print("\nPrediction Result:", result)

if __name__ == "__main__":
    predict_diabetes()

```
